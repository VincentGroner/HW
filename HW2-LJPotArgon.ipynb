```python
from matplotlib import pyplot as plt
import numpy as np
import math 
#the above imports different libraries
%matplotlib inline
#f(r)=4ep((sig/r)^12 - (sig/r)^6)
s = float(3.345)
e = float(125.7) #epsilon/kb from website
x = np.linspace(1,3,60)
y=4*e*((s/x)**12-(s/x)**6)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
#adding figure and adjusting size
ax.plot(x,y)
```
