```python
#Ian and I worked together on this so our code may look quite similar

import pandas as pd
import numpy as np
import math


df = pd.read_csv('MinCoords.csv') #read csv file

i=0
j=0
xi=yi=zi=xj=yj=zj=0
numlines=1350
eps=0.0103 #0.0103eV
sig=0.3345 #also tried 3.4Angs
LJcutoff=3.5
LJ2=LJcutoff**2
EnTot=0

#print(df.to_string()) #test to make sure that I am reading in the correct file
#print(df.iat[0,0]) #test print of a cell
    #[n,0]=atom number
    #[n,1]=atom type (1=argon here)
    #[n,2]=xcoord
    #[n,3]=ycoord
    #[n,4]=zcoord
    #n=ith atom
#print("Atom:", df.iat[0,0], ", Type:", df.iat[0,1], ", X:", df.iat[0,2], ", Y:", df.iat[0,3], ", Z:", df.iat[0,4])

for i in range(0,numlines,1):
    xi=df.iat[i,2]
    yi=df.iat[i,3]
    zi=df.iat[i,4]
    
    for j in range(0, int(i), 1):
        xj=df.iat[j,2]
        yj=df.iat[j,3]
        zj=df.iat[j,4]
        xdif=xj-xi
        ydif=yj-yi
        zdif=zj-zi
        if np.absolute(xdif)<LJcutoff and np.absolute(ydif)<LJcutoff and np.absolute(zdif)<LJcutoff:
        #if xdif2 < LJ2 and ydif2<LJ2 and zdif<LJ2:
            #radius calculation:
            r=np.sqrt(xdif**2+ydif**2+zdif**2)
            if r>0.0 and r<3.5:
                print("i=",i, "AtomI: ", df.iat[i,0], ", X: ", df.iat[i,2], ", Y: ", df.iat[i,3], ", Z: ", df.iat[i,4])
                print("j=",j, "AtomJ: ", df.iat[j,0], ", X: ", df.iat[j,2], ", Y: ", df.iat[j,3], ", Z: ", df.iat[j,4])
                print("radius=", r,",i=",i,"j=",j)
                #energy calculation:
                eg=4*eps*((sig/r)**12-(sig/r)**6)
                #print("Energy=",eg)
                EnTot=EnTot+eg
                #print("EnTot=",EnTot)
    i = i+1
        
    
print("Total Energy:", EnTot)
print("Finished")
    
```
